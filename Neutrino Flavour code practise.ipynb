```python
import numpy as np
from matplotlib import pyplot as plt ##JUNO EQUATION 9.2 
```

```python
def f(p):
    return (1 - (np.cos(0.2))**4*(np.cos(8.57))**4*(np.sin(2*33.44))**2*(np.sin(2.35585*10**(-11)*p))**2
              - (np.cos(0.2))**4*(np.sin(2*8.57))**2*(np.cos(33.44))**2*(np.sin(0.7991475*10**(-9)*p))**2-(np.cos(0.2))**4*(np.sin(2*8.57))**2*(np.sin(33.44))**2*(np.sin(0.775589*10**(-9)*p))**2
              - (np.cos(8.57))**4*(np.sin(2*0.2))**2*(np.cos(33.44))**2*(np.sin(3.175*10**(-6)*p))**2-(np.cos(8.57))**4*(np.sin(2*0.2))**2*(np.sin(33.44))**2*(np.sin(3.1749764415*10**(-6)*p))**2
              - (np.sin(8.57))**2*(np.sin(2*0.2))**2*(np.sin(3.1742008525*10**(-6)*p)))
```

```python
p =  np.linspace(0.1, 100000000000, 1000)
q = f(p)
```

```python
plt.rcParams["figure.figsize"] = [9.50, 4.50]
plt.rcParams["figure.autolayout"] = True


fig, ax = plt.subplots()
ax.plot(p,q, color = 'blue')
ax.set_xlabel('Varying Distances from the Reactor - L (m)',fontsize='12')
ax.set_ylabel('Probability',fontsize='12')

ax.set_xscale('log')
fig.show()
```
