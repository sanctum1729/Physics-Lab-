```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(L):
    return (  - 0.1811534616 * ((np.sin(2.6924*10**(-14)*L))**2) 
              - 0.068294202 * ((np.sin(0.913311*10**(-12)*L))**2)
              - 0.0880387228 *((np.sin(0.8436428572*10**(-12)*L))**2))
```

```python
L = np.linspace(0.1,1000000000000000,1000)
y = f(L)
```

```python
plt.rcParams["figure.figsize"] = [9.50, 4.50]
plt.rcParams["figure.autolayout"] = True


fig, ax = plt.subplots()
ax.plot(L,y)
ax.set_xlabel('Varying Distances from the Reactor - L (m)',fontsize='12')
ax.set_ylabel('Probability',fontsize='12')

ax.set_xscale('log')
fig.show()
```
