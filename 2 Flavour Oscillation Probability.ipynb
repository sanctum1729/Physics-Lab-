```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(l):
    return 0.8*((np.sin(7.42*10**(-5)*l/3.5000000000))**2)
```

```python
h = np.linspace(0.1,10000000,1000) #I NEED TO CONVERT LINSPACE TO LOGSPACE AND LOGARITHMIC SCALE ON X AXIS
y = f(h) #curve should be smooth

plt.rcParams["figure.figsize"] = [9.50, 4.50]
plt.rcParams["figure.autolayout"] = True

#plt.plot (h, y)
#plt.figure(num=0, dpi=420)
#plt.show()


fig, ax = plt.subplots()
ax.plot(h,y)
ax.set_xscale('log')
fig.show()
```

```python
def f(l):
    return (1 - 0.8*((np.sin(7.42*10**(-5)*l/3.5000000000))**2)) #survival probability
```

```python
h = np.linspace(0.1,10000000,1000) #I NEED TO CONVERT LINSPACE TO LOGSPACE AND LOGARITHMIC SCALE ON X AXIS
y = f(h) #curve should be smooth

plt.rcParams["figure.figsize"] = [9.50, 4.50]
plt.rcParams["figure.autolayout"] = True

#plt.plot (h, y)
#plt.figure(num=0, dpi=420)
#plt.show()


fig, ax = plt.subplots()
ax.plot(h,y,color='red')
ax.set_xscale('log')
fig.show()
```
