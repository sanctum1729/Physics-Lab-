```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(h,e):
    return (1 - (np.sin(2*0.2)**2) - ((np.sin(0.5*h/4*e))**2))
```

```python
h = np.linspace(1000.0,100000000000.0,num=1000,dtype='f')
y = f(h,35000000000)
plt.plot (h, y)
plt.figure(num=0, dpi=120)
plt.show()
```
