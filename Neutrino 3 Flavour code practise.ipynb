```python
#p=theta12 
#q=theta13
#m=m21 #n=m31 #o=m32
#L=baseline
#e=energy

def f(L,m,n,p,q,o,e):
   return (1 -((np.sin(2*p)**2))*((np.cos(q))**4)*(np.sin(m**2*L/4*e)) - ((np.sin(2*q)**2))*((np.cos(p))**2*(np.sin(n**2*L/4*e))-((np.sin(2*q)**2))*((np.sin(p))**2)*(np.sin(o**2*L/4*e))))
```
